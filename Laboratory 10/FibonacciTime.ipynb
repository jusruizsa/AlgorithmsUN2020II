#Fibonnaci recursivo


```python
def fibR(n):
    if n==0 :
        return 0
    if n==1 :
        return 1
    return fibR(n-1)+fibR(n-2)
```

```python
import time
import matplotlib.pyplot as plt

times = []
for n in range(0, 51):
  t0 = time.clock()
  k = fibR(n)
  tf =time.clock()
  times.append(tf-t0)

plt.plot(times)
plt.ylabel('recursive time')
plt.show()
```

#Fibonacci iterativo con memoria

```python
def fibM(n):
    memo = {}
    def fib(n):
        if n==0 :
            return 0
        if n==1 :
            return 1
        if (n-2) not in memo:
            memo[n-2] = fib(n-2)
        if (n-1) not in memo:
            memo[n-1] = fib(n-1)
        return memo[n-1] + memo[n-2]
    return fib(n)
```

```python
import time
import matplotlib.pyplot as plt

times = []
for n in range(0, 501):
  t0 = time.clock()
  k = fibM(n)
  tf =time.clock()
  times.append(tf-t0)

plt.plot(times)
plt.ylabel('iterative memory time')
plt.show()
```

#Fibonacci iterativo

```python
def fib(n):
    if n==0 :
        return 0
    a, b = 0, 1
    for i in range(1, n):
        a, b = b, a+b
    return b
```

```python
import time
import matplotlib.pyplot as plt

times = []
for n in range(0, 501):
  t0 = time.clock()
  k = fib(n)
  tf =time.clock()
  times.append(tf-t0)

plt.plot(times)
plt.ylabel('iterative time')
plt.show()
```
